# Exercise 8-1
## Write a program that initially displays the button in Figure with blue text.

When the button is left-clicked with the mouse, the button should appear with black text and the caption replaced with `Change color of Text to Blue`. Subsequent left-clicks should toggle between the two displays. Note: Insert padx=`50` into the grid statement.

```python
from tkinter import *

def change_color():
    if button["fg"] == "blue":
        button.config(text="Change Color of Text to Blue", fg="black")
    else:
        button.config(text="Change Color of Text to Black", fg="blue")

window = Tk()
window.title("Change Colors")

button = Button(
    window, text="Change Color of Text to Black", fg="blue", command=change_color
)
button.grid(row=0, column=0, padx=50, pady=20)

window.mainloop()
```

# Exercise 8-3
## Write a program that requests a person's first and last name and then displays their full name in a read-only box after clicking the `Display Full Name` button.

See Figure.

image.png

```python
from tkinter import *

def display_full_name():
    full_name.set(first_name.get() + " " + last_name.get())

window = Tk()
window.title("Full Name")
window.config(padx=10, pady=5)

last_name = StringVar()
first_name = StringVar()
full_name = StringVar()

last_name_label = Label(window, text="Last name:")
last_name_entry = Entry(window, textvariable=last_name)

first_name_label = Label(window, text="First name:")
first_name_entry = Entry(window, textvariable=first_name)

full_name_btn = Button(window, text="Display Full Name", command=display_full_name)
full_name_label = Label(window, text="Full name:")
full_name_entry = Entry(window, textvariable=full_name, state="readonly")

last_name_label.grid(row=0, column=0, pady=5)
last_name_entry.grid(row=0, column=1, pady=5)
first_name_label.grid(row=1, column=0, pady=5)
first_name_entry.grid(row=1, column=1, pady=5)
full_name_btn.grid(row=2, column=0, columnspan=2, pady=5)
full_name_label.grid(row=3, column=0, pady=5)
full_name_entry.grid(row=3, column=1, pady=5)

window.mainloop()
```
